```python
import pandas as pd
import numpy as np
```

# Hispanic Students

## Load excel file

```python
latino_df = pd.read_excel("acs2019_5yr_B01001I_Hispanic.xlsx", skiprows = 1, header=[0,1], sheet_name=0)
latino_df = latino_df[latino_df.columns.drop(list(latino_df.filter(regex='Error')))]
latino_df.columns = latino_df.columns.droplevel(1)
```

```python
latino_df.head()
```

## drop unwanted zip codes

```python
latino_df.drop(columns=['28754','28705','28622','42602', '28771','28906', '35739'], inplace=True)
latino_df.head()
```

## Transpose data

```python
latino_df_clean = latino_df.set_index('Unnamed: 0_level_0').T.reset_index().rename({'index' : 'zip'}, axis=1)
```

```python
latino_df_clean.head()
```

## Export to Excel

```python
latino_df_clean.to_excel(r'latino_students.xlsx', index = False)
```

## Select Students Only

```python
latino_df_clean_columns = latino_df_clean[['zip',  '5 to 9 years', '10 to 14 years', '15 to 17 years', '18 and 19 years']]
```

```python
latino_df_clean_columns.head()
```

## Add ACS_student_total column

```python
latino_df_clean_columns['ACS_ethnicity_total'] = latino_df_clean_columns.sum(axis=1)
latino_df_clean_columns.head()
```

## Add Ethnicity Column

```python
latino_df_clean_columns['ethnicity'] = 'Hispanic'
latino_df_clean_columns.head()
```

# Calculate Number and % of Other Ethnicity

```python
major_ethnicity_df = combined_ethnicity_merged.groupby('zip').ACS_ethnicity_total.sum().reset_index()
```

```python
total_students_calc_df = combined_ethnicity_merged.groupby('zip').ACS_total_students.mean().reset_index()
```

```python
#other_students_df = pd.merge(major_ethnicity_df, total_students_calc_df, on = 'zip', how = 'left')
```

```python
other_students_df2 = pd.merge(major_ethnicity_df, total_students_calc_df, on = 'zip', how = 'left')
```

```python
other_students_df2.head(1)
```

```python
other_students_df2.rename({'ACS_ethnicity_total' : 'ACS_ethnicity_total_oc', 'ACS_total_students':'ACS_total_students_oc'}, axis=1, inplace=True)
```

```python
other_students_df2.head()
```

```python
other_students_df2.head()
```

```python
other_students_df2['ethnicity'] = 'Other Ethnicity'
```

```python
other_students_df2.head()
```

```python
other_students_df2['ACS_ethnicity_total'] = other_students_df2['ACS_total_students_oc'] - other_students_df2['ACS_ethnicity_total_oc']
```

```python
other_students_df2.head()
```

```python
other_students_df2['pct_ethnicity'] = 1 - other_students_df2['ACS_ethnicity_total_oc'] / other_students_df2['ACS_total_students_oc']
```

```python
other_students_df2.rename({'ACS_total_students_oc':'ACS_total_students'}, axis=1, inplace=True)
```

```python
other_students_df2.head()
```

```python
other_students_df2 = other_students_df2[['zip','ACS_ethnicity_total', 'ethnicity', 'ACS_total_students', 'pct_ethnicity']]
```

```python
other_students_df2.head()
```

```python
combined_ethnicity_merged.info()
```

```python
other_students_df2.info()
```

```python
#other_students_df2.index = list(range(2520,3150))
pd.concat([combined_ethnicity_merged.reset_index(drop = True), other_students_df2], ignore_index = True, sort=False).fillna(0)
```
